# Задание 2

Требуется написать код, рассчитывающий количество рабочих часов с использованием методов модуля datetime.
Требования к коду:
1. Написать функцию *input_date()*, которая запрашивает у пользователя диапазон дат, на котором считается рабочее время. Граничные даты включаются в расчет. Функция возвращает две переменные с датами *start_date, end_date*.
2. Написать функцию *date_to_datetime(start_date, end_date)*, которая преобразует даты в формат datetime. Функция возвращает переменные *start_date, end_date*, преобразованные к формату datetime.
3. Написать функцию *delta_time(start_date, end_date)*, которая рассчитывает разницу во времени. По результату работы функции, в переменную *delta* функции, которая вызвала delta_time, должно быть записано общее количество оцениваемых дней.
4. Написать функцию *day_of_the_week(start_date)*, которая возвращает номер дня недели, соответствующий началу оцениваемого периода, в переменную *start_day_of_the_week*.
5. Написать функцию *calculating_hours(start_day_of_the_week, delta)*, которая возвращает количество рабочих часов за указанный период (и, таким образом, является точкой входа в нашу программу).
6. Сформировать тестовые данные и протестировать программу.

Упрощения:
1. Пятидневная рабочая неделя, суббота и воскресенье - выходные;
2. Каждый день отрабатывается ровно 8 часов;
3. Праздники, отгулы, отпуска и больничные отсутствуют

```python
from datetime import datetime
def input_date():
  while True:
    try:
      dayStart = int(input("Введите день начала работ: "))
      monthStart = int(input("Введите месяц: "))
      yearStart = int(input("Введите год: "))
    except ValueError:
      print("Error! Это не число, попробуйте снова.")
    else:
      break
  while True:
    try:
      dayEnd = int(input("Введите день окончания работ: "))
      monthEnd = int(input("Введите месяц: "))
      yearEnd = int(input("Введите год: "))
    except ValueError:
      print("Error! Это не число, попробуйте снова.")
    else:
      break
  if dayStart < 10:
    strdayStr = '0' + repr(dayStart)
  else:
    strdayStr = repr(dayStart)
  if monthStart < 10:
    strmonStr = '0' + repr(monthStart)
  else:
    strmonStr = repr(monthStart)
  if dayEnd < 10:
    strdayEnd = '0' + repr(dayEnd)
  else:
    strdayEnd = repr(dayEnd)
  if monthEnd < 10:
    strmonEnd = '0' + repr(monthEnd)
  else:
    strmonEnd = repr(monthEnd)
  stryearStr = repr(yearStart)
  if len(stryearStr) < 4:
    while len(stryearStr) != 4:
      stryearStr = '0' + stryearStr
  stryearEnd = repr(yearEnd)
  if len(stryearEnd) < 4:
    while len(stryearEnd) != 4:
      stryearEnd = '0' + stryearEnd
  start_date = stryearStr + '-' + strmonStr + '-' + strdayStr
  end_date = stryearEnd + '-' + strmonEnd + '-' + strdayEnd
  return start_date, end_date

def date_to_datetime(start_date, end_date):
  try:
    con_start_date = datetime.fromisoformat(start_date + ' 00:00:00')
    con_end_date = datetime.fromisoformat(end_date + ' 00:00:00')
  except ValueError:
    print("Error! Неверный формат даты! Создан новый период!")
    con_start_date = datetime.fromisoformat('2022-03-04 00:00:00')
    con_end_date = datetime.fromisoformat('2023-03-04 00:00:00')
    print("Начало работ: ",con_start_date)
    print("Конец работ: ", con_end_date)
    return con_start_date, con_end_date
  print("Начало работ: ",con_start_date)
  print("Конец работ: ", con_end_date)
  return con_start_date,con_end_date

def delta_time(start_date, end_date):
  delta = end_date - start_date
  return delta

def day_of_the_week(start_date):
  return datetime.isoweekday(start_date)

def calculating_hours(start_day_of_the_week, delta):
  i = 0
  j = start_day_of_the_week
  res = 0
  if i > delta:
    print ("Error! Не правельные условия!")
  while i <= delta:
    if j < 6:
      res += 8
    if j == 7:
      j = 0
    i += 1
    j += 1
  return res

def test():
  start_date , end_date = input_date()
  csd,ced = date_to_datetime(start_date,end_date)
  DELTA = delta_time(csd,ced)
  start_day_of_the_week = day_of_the_week(csd)
  print("Период из: ",DELTA.days + 1,"дней.")
  print("Номер дня недели старта периода: ",start_day_of_the_week)
  hours = calculating_hours(start_day_of_the_week, DELTA.days)
  print("В период отработано: ",hours, "часов.")

test()    
```
